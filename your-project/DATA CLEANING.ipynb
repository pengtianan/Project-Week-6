Create a function to approximate the number of rooms for each housing of the dataset using the following classification:



Source: https://eldorado-immobilier.com/definition-t1t2t3t4-studio-quelle-surface-et-combien-de-pieces/
consulted, February 26th

```python
def class_house(area):
    if area < 32:
        return "Studio"
    elif area >= 32 and area < 65:
        return "1"
    elif area >= 65 and area < 80:
        return "2"
    else:
        return "3+"

```

```python

rent_clean["Rooms2"] = rent_clean["Area"].apply(class_house)
rent_clean["Number of bedrooms"] = np.where(rent_clean.Rooms.notna(), rent_clean.Rooms, rent_clean.Rooms2)
rent_clean.drop(["Rooms","Rooms2"],axis=1,inplace=True)
#rent_clean.drop(["Rooms2"],axis=1,inplace=True)
```
